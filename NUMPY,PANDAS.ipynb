```python
x=[[1,2 ,3],[2, 3, 4],[3, 4, 5]]
ar=np.array(x)
ar.sum(axis=1)
print(ar.T)
print(ar.argmax())                #returns index of largest element in array
print(ar.argsort())               #returns indices that sort each row
```
